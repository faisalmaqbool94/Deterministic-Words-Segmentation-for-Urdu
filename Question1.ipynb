```python
with open('dictionary.txt', 'r', encoding='utf-8') as myfile:
  dictionary = myfile.read()
```

```python
tokens = [s.strip() for s in dictionary.splitlines()]
```

```python
result = []
def maxMatchAlgorithm(sentence,wordlist):
    #BAse Check if empty then return and don't add anything in the result list
    if sentence == '':
        return []
    
    #looping from top (lenth of sentence to bottom 0th character) to get the max match
    for i in range(len(sentence)-1,0,-1):
        
        firstWord = sentence[:i+1]      #first i characters of sentense
        remainder = sentence[i+1:]      #rest of the sentense
        
        #check if the firstword in the wordlist
        if firstWord in tokens:
            return [firstWord] + maxMatchAlgorithm(remainder,wordlist)
            
    firstWord = sentence[0:1]
    remainder = sentence[1:]
    
    return [firstWord] + maxMatchAlgorithm(remainder,wordlist)
```

```python
result = maxMatchAlgorithm(data,tokens)
```
